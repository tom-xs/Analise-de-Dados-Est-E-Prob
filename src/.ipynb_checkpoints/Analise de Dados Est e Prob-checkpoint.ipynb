<H1>Regressão Linear e Testes de Hipotese </H1>

Nessa parte tentaremos encontrar uma equação que descreva a nota bruta dos estudantes do curso utilizando o metodo OLS(Ordinary least squares) para regressão linear<p>
<b><font size = "2">a função utilizada para esta analise (<b>OLS</b>) tem como padrão $\alpha_P = 5\%$</font></b>

Inicialmente tentaremos descrever a Nota Bruta a partir das seguintos variáveis:<ul>
    <li>Nota no componente especifico</li>
    <li>Tipo de escola que o aluno cursou no ensino medio</li>
    <li>Numero de pessoas que moram na mesma casa</li>
    <li>Renda familiar total</li>
    <li>Idade</li>
</ul>

```python
resultado = sm.OLS(dados['NOTA BRUTA'],dados[['NOTA COMPONENTE ESPECÍFICO','ESCOLAEM','NPESSOASCASA','RENDAFAMILIA','IDADE']]).fit()
```

```python
print(resultado.summary())
```

Vamos começar comentando os dados da primeira tabela apresentada. As medidas mais importantes para nós neste momento são o R² ajustado(R-squared), a estatística de teste F (F-statisc) e  o p-valor dessa estatística(Prob(F-statistic)) <p>
<ul>
    <li>O valor R² nos mostra o quanto esse modelo explica as variaveis, nesse caso $R² = 99.5\%$, ou seja, o modelo escolhido descreve muito bem a Nota Bruta</li>
    <li>a estatística de teste F e seu p-valor $< 0.001$ basicamente nos mostram que esse modelo é estatisticamente válido</li>
</ul>

analisando os coeficientes obtidos pela função <b>OLS</b> teremos a seguinte equação
$$y = 0.8831x_1-1.2331x_2+0.277x_3-0.0003x_4+0.4032x_5$$
<p>onde:<ul>
    <li>$y$ = Nota Bruta</li>
    <li>$x_1$ = Nota Componente Específico</li>
    <li>$x_2$ = Tipo de escola que cursou no Ensino Médio</li>
    <li>$x_3$ = Número de pessoas que moram na mesma casa</li>
    <li>$x_4$ = Renda total familiar</li>
    <li>$x_5$ = Idade</li>
    

porém, vemos alguns problemas com as variáveis escolhidas devido aos avisos dados na tabela, onde diz que <b>Cond. No.</b> deu um valor muito alto

<font size="2"><b>obs:a função desta variavél é indicar o quão sensível o resultado da função encontrada é em relação aos valores escolhidos para descrever-lá</font></b>

para sabermos quais dados válidos para a nova regressão iremos analisar a coluna $P > |t|$, pois quanto mais proximo seu valor de 0 mais relevante esse coeficiente é no modelo adotado, Logo, em busca do modelo ideal, iremos remover as variaveis que não atendem a esse criterio (ESCOLAEM ,NPESSOASCASA , RENDAFAMILIA)

```python
resultado = sm.OLS(dados['NOTA BRUTA'],dados[['NOTA COMPONENTE ESPECÍFICO','IDADE']]).fit()
```

```python
print(resultado.summary())
```

logo, analisando os novos coeficientes obtidos teremos a seguinte equação<p>
    $$y = 0.8799x_1+0.3511x_2$$
onde<ul>
    <li>$y$ = Nota Bruta</li>
    <li>$x_1$ = Nota Componente Específico</li>
    <li>$x_2$ = Idade</li>
</ul>

Mais uma vez, comentando os dados da primeira tabela apresentada.<p>
<ul>
    <li>mais uma vez $R² = 99.5\%$, ou seja, o modelo continua descrevendo muito bem a Nota Bruta</li>
    <li>a estatística de teste F e seu p-valor $< 0.001$ porém, nesse caso o valor de Prob(f) é ainda menor que o mostrado na primeira análise, o que nos indica que as variáveis usadas nessa nova regressão são ainda mais estatisticamente validos que os anteriores</li>
</ul>
